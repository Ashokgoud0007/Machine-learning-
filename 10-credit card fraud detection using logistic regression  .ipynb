## importing the necessary libruaries 

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

## data collection and preprocessing 

```python
# load the dataset
credit_card_data=pd.read_csv(r"C:\Users\USER\Downloads\creditcard.csv")
```

```python
# check for missing values for dataset
credit_card_data.isnull().sum()
```

```python
# destribution of legit transaction and fraudlent transaction 
credit_card_data['Class'].value_counts()
```

### normal transaction ---->0
### fraudlent transaction --->1

#### this dataset is highly unbalenced

```python
# separating data for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions 
credit_card_data.groupby('Class').mean()
```

### boild a sample dataset containing similar distribution and fraudlent transaction
### num of fraudlent transactions--->492

```python
legit_sample=legit.sample(n=492)
```

### concatinating two dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

### split the data into feature & target

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
x
```

```python
y
```

### split the data into training and testing 

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,
                                                 stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

## model training 

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
# training the logistic regression with training data

x_train_pred=model.predict(x_train)
x_train_pred
```

```python
training_accuracy_score=accuracy_score(x_train_pred,y_train)
```

```python
print('training_accuracy_score:',training_accuracy_score)
```

```python
# testing the logistic regression with testing data


# accuracy on testind data
x_test_pred=model.predict(x_test)
x_test_pred
```

```python
testing_accuracy_score=accuracy_score(x_test_pred,y_test)
```

```python
print('testing_accuracy_score:',testing_accuracy_score)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.pyplot as plt

# Assuming y_train, y_test, y_train_pred, y_test_pred are defined
plt.figure(figsize=(14, 7))

# Plot training predictions
plt.subplot(1, 2, 1)
plt.plot(y_train, label='True Values', color='blue')
plt.plot(x_train_pred, label='Predictions', color='orange')
plt.title('Training Predictions vs True Values')
plt.xlabel('Samples')
plt.ylabel('Values')
plt.legend()

# Plot testing predictions
plt.subplot(1, 2, 2)
plt.plot(y_test, label='True Values', color='blue')
plt.plot(x_test_pred, label='Predictions', color='orange')
plt.title('Testing Predictions vs True Values')
plt.xlabel('Samples')
plt.ylabel('Values')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# residual plot
plt.figure(figsize=(14, 7))

# Residuals for training data
plt.subplot(1, 2, 1)
plt.scatter(x_train_pred, x_train_pred - y_train)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals for Training Data')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')

# Residuals for testing data
plt.subplot(1, 2, 2)
plt.scatter(x_test_pred, x_test_pred - y_test)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals for Testing Data')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')

plt.tight_layout()
plt.show()
```

```python
# scatter plot
plt.figure(figsize=(14, 7))

# Histogram for training errors
plt.subplot(1, 2, 1)
plt.hist(x_train_pred - y_train, bins=5, alpha=0.7, color='blue')
plt.title('Histogram of Training Errors')
plt.xlabel('Error')
plt.ylabel('Frequency')

# Histogram for testing errors
plt.subplot(1, 2, 2)
plt.hist(x_test_pred - y_test, bins=30, alpha=0.7, color='orange')
plt.title('Histogram of Testing Errors')
plt.xlabel('Error')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
